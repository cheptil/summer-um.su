```python
from collections import defaultdict
```

```python
facts = defaultdict(set)
count = 0
f=0
m=0
n=0
with open("Основные списки (25 мая) fm.csv") as ff:
    for i in ff.readlines()[1:]:
        count += 1
        ii = i.rstrip().split(';')
        if ii[1] == 'f':
            f = f + 1
        elif ii[1] == 'm':
            m = m + 1
        elif ii[1] == '?':
            n = n + 1
        facts[ii[6]].add(' '.join(filter(None, ii[:-1])))
print(count, m, f, n)
with open("Резервные списки (25 мая) fm.csv") as ff:
    for i in ff.readlines()[1:]:
        count += 1
        ii = i.rstrip().split(';')
        if ii[1] == 'f':
            f = f + 1
        elif ii[1] == 'm':
            m = m + 1
        elif ii[1] == '?':
            n = n + 1
        facts[ii[6]].add(' '.join(filter(None, ii[:-1])))
print(count, m, f, n)
```

```python
def popper(names):
    global facts
    for name in names[1:]:
        facts[names[0]]=facts[names[0]].union(facts[name])
        print(facts.pop(name))
    return sum(len(facts[i]) for i in facts)
```

```python
import matplotlib.pyplot as plt
from collections import OrderedDict
```

```python
def plotter(title, ylabel, facts, axis, name):
    res=[]
    for fact in sorted(facts.keys()):
        res.append((fact, len(facts[fact])))
    
    count=0
    for i,j in sorted(res, key=lambda x: x[1], reverse=True):
        count += j
        print(i, j)
    print(len(facts.keys()))
    print(count)

    #fa = {i:len(j) for i,j in facts.items()}
    fa = defaultdict(lambda: [0, 0, 0])
    for i,k in facts.items():
        for j in k:
            if j.split(' ')[1] == 'm':
                fa[i][0] += 1
            elif j.split(' ')[1] == 'f':
                fa[i][1] += 1
            elif j.split(' ')[1] == '?':
                fa[i][2] += 1
 
    fa=OrderedDict(sorted(fa.items(), key=lambda t: t[0]))
    print(fa)
    print(len(fa.keys()))  
    print(sum(sum(fa[i]) for i in fa))
    
    m = [i[0] for i in fa.values()]
    f = [i[1] for i in fa.values()]
    o = [i[2] for i in fa.values()]
    
    plt.figure(figsize=(15,8))
    p1 = plt.bar(fa.keys(), m)
    p2 = plt.bar(fa.keys(), f, bottom=m)
    p3 = plt.bar(fa.keys(), o, bottom=[x + y for x, y in zip(m, f)])
    plt.xticks(list(fa.keys()), list(fa.keys()), rotation='vertical')
    plt.ylabel(ylabel, size=17)
    plt.title(title, size=17)
    plt.axis(axis)
    if sum(o):
        plt.legend((p1[0], p2[0], p3[0]), ('Men', 'Women', 'Not known'), fontsize=17, loc=[0.85, 0.8])
    else:
        plt.legend((p1[0], p2[0]), ('Men', 'Women'), fontsize=17, loc=[0.85, 0.8])
    plt.tick_params(labelsize=17)
    #plt.figure(figsize=(10,60))
    #plt.show()
    plt.savefig(name, bbox_inches = 'tight')
```

```python
plotter('Распределение участников спартакиады по факультетам', 'Количество участников спартакиады', facts, [-1, 31, 0, 120], 'facts.png')
```

```python
plotter('Распределение участников спартакиады по курсам', 'Количество участников спартакиады', years, [-1, 8, 0, 250], 'years.png')
```
